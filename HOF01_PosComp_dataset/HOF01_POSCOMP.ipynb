```python
def get_github_raw_link(github_link):
    """
    Converte um link de arquivo do GitHub (visualização de blob)
    em um link raw para download direto.
    """
    raw_link = github_link.replace("https://github.com/", "https://raw.githubusercontent.com/")
    raw_link = raw_link.replace("/blob/", "/")
    return raw_link

# Exemplo com o link do seu dicionário 'poscomp_fundamentos'
original_link = poscomp_fundamentos_blob["2002"]
raw_download_link = get_github_raw_link(original_link)

print(f"Link Original: {original_link}")
print(f"Link para Download Raw: {raw_download_link}")
```

```python
!pip install requests pdfplumber

import requests
import pdfplumber
import io

# Dicionário para armazenar o texto extraído de cada PDF
poscomp_texts = {}

# Função para baixar e extrair texto de um PDF
def download_and_extract_pdf_text(year, blob_url):
    raw_url = get_github_raw_link(blob_url)
    print(f"Tentando baixar {year} de: {raw_url}")
    try:
        response = requests.get(raw_url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx)

        # Usa io.BytesIO para tratar o conteúdo como um arquivo em memória
        with pdfplumber.open(io.BytesIO(response.content)) as pdf:
            all_text = []
            for page in pdf.pages:
                text = page.extract_text()
                if text:
                    all_text.append(text)
            print(f"PDF {year} baixado e texto extraído com sucesso.")
            return "\n".join(all_text)
    except requests.exceptions.RequestException as e:
        print(f"Erro ao baixar o PDF {year} de {raw_url}: {e}")
        return None
    except pdfplumber.PDFSyntaxError as e:
        print(f"Erro de sintaxe no PDF {year}: {e}")
        return None
    except Exception as e:
        print(f"Um erro inesperado ocorreu ao processar o PDF {year}: {e}")
        return None

# Itera sobre o dicionário de links e processa cada PDF
for year, blob_url in poscomp_fundamentos_blob.items():
    extracted_text = download_and_extract_pdf_text(year, blob_url)
    if extracted_text:
        poscomp_texts[year] = extracted_text

print("\nProcessamento concluído.")
print(f"PDFs processados: {list(poscomp_texts.keys())}")
```

```python
import re

# 0. Select Year (2004)
index = 2
listaAnos = list(poscomp_fundamentos_blob.keys())
listaURL = list(poscomp_fundamentos_blob.values())
tuplaYearUrl = [listaAnos[index], listaURL[index]]

ocr_text = extract_ocr_page_3(tuplaYearUrl[0], tuplaYearUrl[1])
print(f"Processing: {tuplaYearUrl[0]} (Page 3)\n")

# Refined Regex: Look for numbers 1-70 followed by dot/space.
# We look for double spaces or newlines to identify a new block.
question_split_pattern = r'(?:^|\n|\s{2,})(\b(?:[1-9]|[1-6][0-9]|70)\s*[\.\-]?\s+)(?=[A-Z\u00C0-\u00DC])'

structured_questions = []
parts = re.split(question_split_pattern, ocr_text)

i = 1
while i < len(parts):
    num_raw = parts[i].strip()
    num = re.sub(r'[^0-9]', '', num_raw)
    content = parts[i+1] if i+1 < len(parts) else ""

    # Clean leakage from the next question marker
    clean_content = re.split(r'\s+\d+\s+[A-Z\u00C0-\u00DC]', content)[0]

    # Extract alternatives using a pattern that looks for (a) or a)
    alts = re.findall(r'(\(?[a-e][\)\.]\s*[^\(]+)', clean_content)

    # Fallback: If 'a' is missing but 'b' is present, it's often merged in the statement
    if not any(a.strip().startswith(('a', '(a')) for a in alts):
        first_alt_match = re.search(r'(?:\?|:)\s*([^\(]+?)\s*\(b\)', clean_content)
        if first_alt_match:
            alts.insert(0, f"(a) {first_alt_match.group(1)}")

    # Statement: everything before the first actual alternative marker
    # This time we strictly split and keep only the first part to avoid duplicating alts in the statement
    split_stmt = re.split(r'\s*\(?[a][\)\.]|\s+[a][\)\.]', clean_content, maxsplit=1)
    statement = split_stmt[0].strip()

    # Special case: if (a) was '6' and merged, remove it from statement
    if alts and "(a)" in alts[0]:
        val_a = alts[0].replace("(a)", "").strip()
        if statement.endswith(val_a):
            statement = statement[:statement.rfind(val_a)].strip()

    if len(statement) > 5:
        structured_questions.append({
            "number": num,
            "statement": statement,
            "alternatives": [a.strip() for a in alts]
        })
    i += 2

# Final Output
for q in structured_questions:
    print(f"QUESTION {q['number']}")
    print(f"Statement: {q['statement']}")
    if q['alternatives']:
        for alt in q['alternatives']:
            print(f"  {alt}")
    else:
        print("  [No alternatives detected]")
    print("-" * 50)
```
